```python
import re
```

```python
# Define a function that takes the file path as input
def log_analyzer(file_path):
    # Create an empty dictionary to store log counts
    # Example: {"ERROR": 2, "INFO": 1}
    log_count={}
    # Open the file in read mode ("r")
    # 'with' ensures the file is automatically closed after use
    with open('logfile.txt',mode='r') as file:
        for lines in file:
                # Search for a pattern of uppercase words (like ERROR, INFO, WARNING)
            match = re.search(r'\b(INFO|ERROR|WARNING|DEBUG|CRITICAL)\b', lines)
            if match:
                # Extract the matched word (e.g., "ERROR")
                var=match.group()
                # Update dictionary count:
                # If key exists → increment count
                # If not → initialize with 0 and then add 1
                log_count[var]=log_count.get(var,0)+1
    return log_count
            
```

```python
a=log_analyzer('logfile.txt')
#Calculate total number of logs
total=sum(a.values())

for key,value in sorted(a.items()):
     # Calculate percentage of each log type
    percentage=(value/total)*100
    # Print log type, count, and percentage (formatted to 2 decimal places)
    print(f'{key}:{value} ({percentage:.2f}%)')
  
    
    
```

```python

thresholds = {
    "ERROR": 10 ,#%
    "WARNING": 15 #%
}
for key,value in sorted(a.items()):
    percentage=(value/total)*100
    #
    if key in thresholds and percentage > thresholds[key]:
        print(f" ALERT: {key} too high!")
```
